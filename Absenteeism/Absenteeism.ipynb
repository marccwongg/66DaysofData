```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv('Absenteeism-data.csv')
raw_data.head()
```

```python
df = raw_data.copy()
df.head()
```

```python
df.info()
```

```python
#drop label data
df = df.drop(['ID'], axis=1)
df.head()
```

```python
#create dummy variables
#drop reason 0 to prevent multicollinearity
reason_col = pd.get_dummies(df['Reason for Absence'], drop_first=True)
reason_col
```

```python
#drop Reason from main df
df = df.drop(['Reason for Absence'], axis=1)
df.head()
```

```python
#cetgorise reaason based on respective type
reason_type1 = reason_col.loc[:,1:14].max(axis=1)
reason_type2 = reason_col.loc[:,15:17].max(axis=1)
reason_type3 = reason_col.loc[:,18:21].max(axis=1)
reason_type4 = reason_col.loc[:,22:].max(axis=1)
```

```python
#concat main df and reason type
df = pd.concat([df, reason_type1, reason_type2, reason_type3, reason_type4], axis=1)
df
```

```python
#rename column names
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason1', 'Reason2', 'Reason3', 'Reason4']
```

```python
df.columns = column_names
df.head()
```

```python
#reorder column names
column_names_reordered = ['Reason1', 'Reason2', 'Reason3', 'Reason4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df = df[column_names_reordered]
df.head()
```

```python
#Checkpoint
df_reason_mod = df.copy()
```

```python
#convert Date to timestamp
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format='%d/%m/%Y')
df_reason_mod.head()
```

```python
#Extract month value

months_list = []

for i in range(df_reason_mod.shape[0]):
    months_list.append(df_reason_mod['Date'][i].month)
    
len(months_list)
```

```python
df_reason_mod['Month Value'] = months_list
df_reason_mod.head()
```

```python
#Extract Day of the week
def datetoweekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of The Week'] = df_reason_mod['Date'].apply(datetoweekday)

df_reason_mod.head()
```

```python
#Drop date and reorder
df_reason_mod = df_reason_mod.drop(['Date'], axis=1)
```

```python
df_reason_mod.columns.values
```

```python
column_names_reordered = ['Reason1', 'Reason2', 'Reason3', 'Reason4', 'Month Value',
       'Day of The Week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']

df_reason_mod = df_reason_mod[column_names_reordered]
df_reason_mod.head()
```

```python
#create new checkpoint
df_reason_date_mod = df_reason_mod.copy()
```

```python
#Modify Education column
df_reason_date_mod['Education'].unique()

df_reason_date_mod['Education'].value_counts()
```

```python
df_reason_date_mod['Education'] = df_reason_date_mod['Education'].map({1:0, 2:1, 3:1, 4:1})

df_reason_date_mod['Education'].value_counts()
```

```python
#Create checkpoint for preprocessed df
df_preprocessed = df_reason_date_mod.copy()
```

```python
#Save preprocessed data to new csv file
df_preprocessed.to_csv('Absenteeism_preprocessed.csv', index=False)
```
